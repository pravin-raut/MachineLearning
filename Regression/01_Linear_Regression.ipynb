##Removing Unwanted Columns i.e Car_ID and CarName

```python
df_car_details=data.drop(["car_ID","CarName"],axis=1)
df_car_details
```

```python
cat_col=[]
num_col=[]

for col in df_car_details:
  if df_car_details[col].dtype=='O':
    cat_col.append(col)
  else:
    num_col.append(col)
```

```python
cat_col
```

```python
df_cat=pd.DataFrame(data=df_car_details,columns=cat_col)
df_cat
```

```python
df_num=pd.DataFrame(data=df_car_details,columns=num_col)
df_num
```

```python
for cols in df_cat:
  print(f"{cols} contains {df_cat[cols].nunique()} unique labels")
```

```python
df_cat_encoding=pd.get_dummies(data=df_cat,drop_first=True)
df_cat_encoding
```

```python
df_car_Complete=pd.concat([df_num,df_cat_encoding],axis=1)
df_car_Complete
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error
```

```python
X=df_car_Complete.drop("price",axis=1)
y=df_car_Complete["price"]
```

```python
X_train,X_test,y_train,y_test=train_test_split( X, y, random_state=42, test_size=0.2)
```

```python
column_names = X.columns
column_names_list = list(X.columns)
```

```python
scaler=StandardScaler()

X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
#initializing the model
model = LinearRegression()
```

```python
model.fit(X_train_scaled, y_train)
```

The array slope now holds the coefficients of the linear regression model, in the same order as the features were given when the model was trained. For instance, slope[0] would be the coefficient for the first feature, slope[1] would be the coefficient for the second feature, and so on.

Each coefficient tells you how much the dependent variable is expected to increase when that feature increases by one, holding all other features constant. For example, if slope[0] is 2.5, that means that the model predicts that the dependent variable will increase by 2.5 for every one unit increase in the first feature, assuming all other features remain constant.

The coefficient for wheelbase is positive (643.679737), indicating that as the wheelbase increases, the car's price also tends to increase, all else being equal.

On the other hand, the coefficient for carlength is negative (-989.635897), suggesting that as the car length increases, the car's price tends to decrease, all else being equal. This does not mean car length does not matter. Instead, it suggests that longer cars, according to the model and data, tend to be cheaper.


```python
#To see the simple linear regression parameters
slope = model.coef_
df_coef = pd.DataFrame(slope, index=column_names_list, columns=['Coefficient'])

print(df_coef)

```

The y-intercept (or just "intercept") is still the value that we predict when all our inputs are zero. However, in many real-world situations (like predicting car prices), it doesn't make sense to set all inputs to zero (we can't have a car with zero length, zero weight, etc.), so the intercept may not have a clear real-world interpretation.



```python
y_intercept = model.intercept_
y_intercept
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
mse=mean_squared_error(y_test,y_pred)#mse
r2__score=r2_score(y_test,y_pred)#r2_score
mae=mean_absolute_error(y_test,y_pred)#mae
rmse=mean_squared_error(y_test,y_pred,squared=False)#rmse

#print the metrics 
print("The Metrics for {}:".format(model))
print("----------------------------")
print("Mean Squared Error:{:.2f}".format(mse))
print("Root Mean Squared Error:{:.2f}".format(rmse))
print("Mean Absolute Error:{:.2f}".format(mae))
print("r2_score:{:.2f}".format(r2__score))
```
